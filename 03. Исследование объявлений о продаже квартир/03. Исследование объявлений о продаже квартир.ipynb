## Предобработка данных

### Стиль заголовков

Для начала исправим стиль в заголовках столбцов `cityCenters_nearest`, `parks_around3000`, `ponds_around3000`.

```python
# переименование столбцов
data = data.rename(columns = {'cityCenters_nearest':'city_сenters_nearest', 
                              'parks_around3000':'parks_around_3000', 
                              'ponds_around3000':'ponds_around_3000'}) 
```

Проверим результат.

```python
data.columns # перечень названий столбцов
```

### Работа с пропусками и аномалиями

Теперь посчитаем, сколько в таблице пропущенных значений.

```python
data.isna().sum() # подсчёт пропусков
```

Также посчитаем пропуски в процентах.

```python
# подсчёт пропусков в процентах
pd.DataFrame(data.isna().mean().round(4)*100).sort_values(by=0)[::-1].style.background_gradient('coolwarm')
```

В некоторых столбцах очень много пропущенных значений. Однако не все пригодятся нам для анализа. Стоит пройтись по каждому столбцу в отдельности, заодно проверяя данные на аномалии. В этом нам поможет таблица с описательной статистикой, которая была выведена на предыдущем этапе.

#### total_images

Число фотографий в объявлении от 0 до 50, целочисленные, пропусков нет. 

#### total_area

Значения площади колеблются от 12 до 900 кв.м., пропусков нет.

#### floor

Этаж квартиры принимает значения от 1 до 33, пропусков нет, целые числа.

#### studio и open_plan

Являются булевым тип данных, пропусков нет.

#### airports_nearest, parks_around_3000, parks_nearest, ponds_around_3000 и ponds_nearest

В этих столбцах много пропусков. Большая часть из них легко объяснима: в маленьком населенном пункте расстояние до аэропорта не имеет значения, может не быть парка или водоема, или они находятся на расстоянии более 3 км. Последнее актуально и для Санкт-Петербурга, к тому же данные могут быть не указаны. Поскольку в дальнейшем исследовании эти колонки нам не понадобятся, оставим всё как есть.

#### is_apartment

Очень много пропусков, булев тип данных. Скорее всего, если значение не заполнено, то подразумевается `False`. Проставим `False` в пропусках.

```python
data['is_apartment'] = data['is_apartment'].fillna(False) # заполнение пропусков
```

#### balcony

Число балконов в квартире от 0 до 5, много незаполненных значений, тип `float`. Нужно преобразовать в `int` и заполнить пропуски. Поскольку этот столбец не пригодится нам для дальнейшего анализато предположим, что если данных нет, значит нет и балконов, заменим пропуски на 0.

```python
data['balcony'] = data['balcony'].fillna(0) # заполнение пропусков
data['balcony'] = data['balcony'].astype('int') # изменение типа данных 
```

#### last_price

Большой разброс цен от 12 190 руб. до 763 000 000, дробные числа, пропусков нет.

Во-первых, ни к чему хранить цены на недвижимость в дробных числах. Изменим тип данных на целочисленный.

```python
data['last_price'] = data['last_price'].astype('int') # изменение типа данных 
```

Во-вторых, очень странно выглядит цена в 12 190 руб. для квартиры. Нужно посмотреть, одно ли у нас такое выбивающееся значение, или есть ещё. Посмотрим на 5 объявлений с самой маленькой ценой.

```python
data.sort_values(by='last_price').head() # вывод объявлений с маленькой ценой
```

Вряд ли 12 190 руб. будет стоить квартира в Санкт-Петербурге площадью 109 кв.м. Скорее всего, цена указана не в рублях, а в тысячах рублей. Что касается следующих значений, они выглядят вполне реальными.

Заменим значение 12 190 на 12 190 000.

```python
data['last_price'] = data['last_price'].replace(12190, 12190000) # замена значения
```

#### first_day_exposition

В дате публикации объявления пропусков нет, значения нужно привести к формату `datetime`.

```python
#перевод в формат datetime
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format = '%Y-%m-%dT%H:%M:%S') 
```

Посмотрим, объявления за какой период есть в нашей выборке.

```python
data.agg({'first_day_exposition':['min','max']}) # вывод минимального и максимального значения
```

#### rooms

Принимает значения от 0 до 19, целочисленные, пропусков нет. Во-первых, выглядят подозрительно большие значения. Во-вторых, нужно проверить, являются ли нулевые значения ошибками или нет. Выведем список уникальных значений.

```python
data['rooms'].value_counts().sort_index() # вывод количества уникальных значений
```

Посмотрим, что представляют собой объявления с числом комнат более 10.

```python
data.query('rooms > 10') # срез данных с числом комнат > 10
```

Судя по общей площади, с большим количеством комнат ошибки нет.

Теперь к нулевым значениям. Они могли появиться, если это квартира-студия или квартира со свободной планировкой. Проверим, так ли это в нашей выборке.

```python
# подсчет кол-ва в срезе данных с числом комнат = 0, не являющихся студией или квартирой со свободной планировкой
data.query('rooms == 0 and not (studio == True or open_plan == True)')['rooms'].count()
```

Как видно, в наших данных нет объявлений, где количество комнат равно 0 и при этом они не являются квартирой-студией или квартирой со свободной планировкой. Теперь посмотрим, всегда ли у студии число комнат равно 0.

```python
data.query('studio == True')['rooms'].value_counts() # вывод количества уникальных значений
```

В 11 объявлениях со студиями указана одна комната. Заменим на 0.

```python
data.loc[data['studio'] == True,'rooms'] = 0 # замена значений
data.query('studio == True')['rooms'].value_counts() # проверка
```

#### locality_name

Текстовые данные, 49 пропусков. Посмотрим, какие значения встречаются в этом столбце.

Если пропуски есть в студии или квартире со свободной планировкой, то их надо заменить на 0.

```python
# заполнение пропущенных значений по условию
data.loc[(data['kitchen_area'].isna()) & 
         ((data['studio'] == True) | (data['open_plan'] == True)), 'kitchen_area'] = 0

# подсчет пропусков 
data['kitchen_area'].isna().sum()
```

Для того чтобы заменить оставшиеся пропуски, рассчитаем медианное отношение площади кухни к общей площади квартиры. Стоит отметить, что для квартир с разным числом комнат это отношение должно меняться. Учтем это в расчете.

```python
# расчет медианного отношения площади кухни к общей площади квартиры с группировкой по кол-ву комнат
kitchen_ratio_median = (
    data.groupby('rooms')['kitchen_area'].transform('median') / 
    data.groupby('rooms')['total_area'].transform('median')
)

# заполнение пропущенных значений 
data['kitchen_area'] = data['kitchen_area'].fillna(kitchen_ratio_median * data['total_area']).round(2)

# проверка
data['kitchen_area'].isna().sum()
```

#### living_area

Заполним пропуски в этом столбце так же, как и в площади кухни.

```python
# расчет медианного отношения жилой площади к общей площади квартиры с группировкой по кол-ву комнат
living_ratio_median = (
    data.groupby('rooms')['living_area'].transform('median') / 
    data.groupby('rooms')['total_area'].transform('median')
)

# заполнение пропущенных значений 
data['living_area'] = data['living_area'].fillna(living_ratio_median * data['total_area']).round(2)

# проверка
data['living_area'].isna().sum()
```

```python
data['ceiling_height'].value_counts().sort_index().head(15) # вывод количества самых маленьких уникальных значений
```

По стандарту высота потолков в жилом помещении должна быть не менее 2,5 метров. Но в нашей выборке довольно много значений меньше этого. Самая часто встречающаяся высота для низких потолков - 2,4 метра. Сделаем это значение минимальным.

```python
data.loc[data['ceiling_height'] < 2.4, 'ceiling_height'] = 2.4 # замена значений по условию
```

```python
data['ceiling_height'].value_counts().sort_index().tail(15) # вывод количества самых больших уникальных значений
```

Потолки до 8 метров могут быть например, в многоуровневых или люксовых квартирах. Всё, что выше, выглядит сомнительно. Группа значений от 20 до 32 метров, похоже, указаны в дециметрах. Разделим значения от 24 (так как решили, что это наша минимальная высота) до 32 на 10. Остальные объявления с потолками выше 8 метров удалим.

```python
# деление значений по условию
data.loc[(data['ceiling_height'] <= 32) & (data['ceiling_height'] >= 24), 'ceiling_height'] = data['ceiling_height'] / 10

# удаление строк с высотой потолков от 8 м
data = data.loc[(data['ceiling_height'] < 8.0) | (data['ceiling_height'].isna())]
```

Теперь перейдем к обработке пропусков. Логичнее всего заменить их модой, т.е. самым часто встречающимся значением в выборке.

```python
# расчет моды
ceiling_height_mode = data['ceiling_height'].mode()
print('Мода:' , ceiling_height_mode[0])

# заполнение пропущенных значений 
data['ceiling_height'] = data['ceiling_height'].fillna(ceiling_height_mode[0])
```

### Категоризация этажей

Для покупателей обычно имеет значение, на каком этаже находится квартира: на первом, последнем или другом. Поэтому добавим столбец, который будет охарактеризует расположение квартиры в этом разрезе.

```python
# Функция для категоризации этажей
def floor_category(row):
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] == row['floors_total']:
        return 'последний'
    return 'другой'

# добавление столбца с категориями в data
data['floor_category'] = data.apply(floor_category, axis=1) 

# вывод количества уникальных значений
data['floor_category'].value_counts() 
```

### Выделение центральной зоны Санкт-Петербурга

Посмотрим, какую часть Санкт-Петербурга можно считать центральной, основываясь на стоимости квартир. Для этого для объявлений в Санкт-Петербурге переведем расстояние до центра в километры, для каждого километра найдем среднюю цену и построим график.

```python
# пересчет расстояния до центра в километрах
data_upd['city_сenters_nearest_km'] = (data_upd['city_сenters_nearest'] / 1000).round() 

# расчет средней цены для каждого километра
spb_city_сenters_group = (data_upd.loc[data_upd['locality_name'] == 'Санкт-Петербург']
                          .groupby('city_сenters_nearest_km')['last_price']
                          .mean())

# построение графика
spb_city_сenters_group.plot(x='city_сenters_nearest_km', 
                            y=('last_price', 'mean'), 
                            style='o-', 
                            figsize=(9,5), 
                            grid=True)

plt.title('Зависимость цены от расстояния до центра')
plt.xlabel('расстояние до центра, км')
plt.ylabel('цена, руб.');
```

Как можно увидеть на графике, на расстоянии от 0 до 8 километров до центра квартиры стоят существенно дороже, чем после 8-го километра. Таким образом, можно считать центральной частью Санкт-Петербурга радиус в 8 километров.
