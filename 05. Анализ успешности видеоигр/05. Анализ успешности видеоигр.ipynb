**Цель исследования**: выявить определяющие успешность игры закономерности, чтобы сделать ставку на нужные продукты.

Из открытых источников нам доступны исторические данные до 2016 года о продажах игр, оценки пользователей и экспертов, жанры и платформы.

##  Подготовка данных

###  Стиль заголовков

В названиях колонок встречаются заглавные буквы, что является нарушением стиля. Приведем все названия к нижнему регистру.

```python
data.columns = data.columns.str.lower() # приведение названий столбцов к нижнему регистру
```

###  Изменение типов данных

Нам нужно изменить тип данных для оценок пользователей на числовой, а также привести года к целым числам. 

Среди оценок пользователей встречаются строковые значения `tbd`. Они означают, что оценка "подлежит определению", т.е. по сути это неизвестные нам значения или пропуски. Воспользовавшись методом `pd.to_numeric` с параметром `errors='coerce'` мы одновременно и изменим тип данных, и заменим эти значения на NaN.

```python
data['user_score'] = pd.to_numeric(data['user_score'], errors='coerce') # приведение значений к числовому типу
```

Столбцом `year_of_release` займемся чуть позже, поскольку сначала попробуем восстановить некоторые пропуски.

###  Обработка пропусков

Сначала посчитаем, сколько всего в таблице пропущенных значений, а также их долю от общего числа пропусков.

```python
data.isna().sum() # подсчёт пропусков
```

```python
data.isna().mean().round(4)*100 # подсчёт доли пропусков
```

Обнаружены пропущенные значения в столбцах с названием игры, годом выпуска, оценкой критиков, оценкой пользователей и рейтингом. Причем у последних трех столбцов очень большая доля пропусков.

Посмотрим на пропуски в столбце `name`.

```python
data[data['name'].isna()]  # просмотр строк с пропущеными значениями
```

Пропуски в столбцах `name` и `genre` оказались в одних и тех же строках. Кроме того, здесь же есть пропуски в оценках пользователей и критиков и рейтинге. Эти строки можно смело удалить.

```python
data.dropna(subset = ['name'], inplace=True) # удаление строк с пропусками в name
```

Также стоит привести названия игр к нижнему регистру.

```python
data['name'] = data['name'].str.lower() # приведение к нижнему регистру
```

Перейдем к столбцу `year_of_release`. Для начала посмотрим, за какой период у нас есть данные.

```python
data.agg({'year_of_release':['min','max']}) # вывод минимального и максимального значения
```

Бывает, что в названии игры присутствует год, особенно это касается спортивных серийных игр. Посмотрим, не удастся ли нам заполнить некоторые пропуски, опираясь на этот факт. Сделаем срез игр, у которых в названии есть год, при этом значение в столбце `year_of_release` пропущено.

```python
# получаем срез игр
data.loc[(data['year_of_release'].isna()) & 
         ((data['name'].str.contains('198')) | 
          (data['name'].str.contains('199')) | 
          (data['name'].str.contains('200')) | 
          (data['name'].str.contains('201')))]['name'].sort_values()
```

Соберем серии, в которую входят полученные игры, в одну таблицу, и выведем для каждой игры список уникальных значений для годов выпуска и рейтинга.

```python
# создание списка с сериями игр
series_games = ['all-star baseball', 
                'fifa soccer', 
                'football manager', 
                'madden nfl', 
                'nascar thunder', 
                'nba live', 
                'nfl gameday', 
                'pdc world championship darts', 
                'pro evolution soccer', 
                'sega rally', 
                'tomb raider', 
                'tour de france', 
                'wwe smackdown vs. raw']

# создание пустого датафрейма
data_series_games = pd.DataFrame()

# заполнение датафрейма играми из серий
for game in series_games:
    data_series_games = data_series_games.append(data.loc[data['name'].str.contains(game)])

# снятие ограничений на количество выводимых строк    
pd.set_option('display.max_rows', None)   

# группировка таблицы
data_series_games.groupby('name').agg({'rating':'unique', 'year_of_release':'unique'})
```

Из таблицы видно, что большинство игр выходит на год раньше даты, указанной в названии. Исключение составляют PDC World Championship Darts и Tour de France, у них года совпадают. Для серий Pro Evolution Soccer, Sega Rally и Tomb Raider нельзя сделать однозначных выводов. 

Также можно заметить, что все серии, кроме Tomb Raider, имеют единый рейтинг, причем большинство - Е. У игр PDC World Championship Darts рейтинг E10+, а у WWE SmackDown vs. RAW рейтинг Т.

Сделаем замены пропусков с учетом полученных выводов.

```python
# заполнение пропущенных значений по условию
data.loc[(data['year_of_release'].isna()) & 
         (data['name'] == 'all-star baseball 2005'), 'year_of_release'] = 2004

data.loc[(data['year_of_release'].isna()) & 
         ((data['name'] == 'fifa soccer 2004') | (data['name'] == 'madden nfl 2004')), 'year_of_release'] = 2003

data.loc[(data['year_of_release'].isna()) & 
         ((data['name'] == 'football manager 2007')| (data['name'] == 'madden nfl 07')), 'year_of_release'] = 2006

data.loc[(data['year_of_release'].isna()) & 
         ((data['name'] == 'nascar thunder 2003') | (data['name'] == 'nba live 2003')| 
          (data['name'] == 'nfl gameday 2003')), 'year_of_release'] = 2002

data.loc[(data['year_of_release'].isna()) & 
         (data['name'] == 'pdc world championship darts 2008'), 'year_of_release'] = 2008

data.loc[(data['year_of_release'].isna()) & 
         (data['name'] == 'tour de france 2011'), 'year_of_release'] = 2011

data.loc[(data['year_of_release'].isna()) & 
         ((data['name'] == 'wwe smackdown vs. raw 2006') | (data['name'] == 'madden nfl 06')), 'year_of_release'] = 2005

data.loc[(data['year_of_release'].isna()) & 
         (data['name'] == 'madden nfl 11'), 'year_of_release'] = 2010

data.loc[(data['year_of_release'].isna()) & 
         (data['name'] == 'madden nfl 2002'), 'year_of_release'] = 2001

data.loc[(data['rating'].isna()) & 
         (data['name'].str.contains('pdc world championship darts')), 'rating'] = 'E10+'

data.loc[(data['rating'].isna()) & 
         (data['name'].str.contains('wwe smackdown vs. raw')), 'rating'] = 'T'

series_games_e = ['all-star baseball', 
                  'fifa soccer', 
                  'football manager', 
                  'madden nfl', 
                  'nascar thunder', 
                  'nba live', 
                  'nfl gameday', 
                  'pro evolution soccer', 
                  'sega rally', 
                  'tour de france']

for game in series_games_e:
    data.loc[(data['rating'].isna()) & 
             (data['name'].str.contains('game')), 'rating'] = 'E'

# подсчёт пропусков
data['year_of_release'].isna().sum()
```

Осталось 252 строки с неизвестным годом выпуска. Поскольку это очень важный параметр для нашего дальнейшего исследования, а доля пропусков не очень велика, удалим эти строки.

```python
data.dropna(subset = ['year_of_release'], inplace=True) # удаление строк с пропусками в 'year_of_release'
```

Теперь ничего не мешает привести тип данных в `year_of_release` в целочисленный.

```python
data['year_of_release'] = data['year_of_release'].astype('int') # приведение к целочисленному типу
```

В столбцах `critic_score` и `user_score` находятся оценки критиков и пользователей. Заменить пропуски в этих столбцах не представляется возможным: любая попытка подставить среднее или медианное значение сильно исказит результаты исследования. Поэтому оставим все как есть.

В столбце `rating` наилучшим решением будет замена пропущенных значений на `unknown`.

```python
data['rating'] = data['rating'].fillna('unknown') # замена пропущенных значений на 'unknown'
```

### Изучение влияния оценок пользователей и критиков на продажи

На примере PlayStation 4 посмотрим, как влияют на продажи игр оценки пользователей и критиков. Построим диаграмму рассеяния и посчитаем корреляцию.

```python
# создание среза для платформы PS4 с непустыми значениями оценок
PS4 = data_act.query('platform == "PS4"').dropna()
```

```python
# создание функции для построения диаграммы рассеяния и расчета коэффицента корреляций
def plot_corr(platform):

    fig = plt.figure(figsize=(14,5))
    ax = fig.add_subplot(121)
    sns.scatterplot(data=platform, x='critic_score', y='total_sales')
    plt.title('Зависимость продаж от оценок критиков')
    plt.xlabel('оценки критиков')
    plt.ylabel('продано, млн копий')

    ax = fig.add_subplot(122)
    sns.scatterplot(data=platform, x='user_score', y='total_sales')
    plt.title('Зависимость продаж от оценок пользователей')
    plt.xlabel('платформа')
    plt.ylabel('продано, млн копий')
    plt.show()
    
    print(f'Корреляция с оценками критиков: {platform["total_sales"].corr(platform["critic_score"]).round(3)}')
    print(f'Корреляция с оценками пользователей: {platform["total_sales"].corr(platform["user_score"]).round(3)}')
```

```python
# применение функции для PlayStation 4
plot_corr(PS4)
```

Наблюдается слабая прямая связь между оценками критиков и объемом продаж, а вот оценки пользователей на продажи не влияют. Таким образом, можно сделать вывод, что при покупке игры люди в какой-то мере обращают внимание на мнение критиков. Чем выше оценки критиков, тем больше игра продается.

Проверим, соотносится ли этот вывод с играми на двух других потенциально прибыльных платформах.

```python
# создание среза для платформы Xbox One с непустыми значениями оценок
XOne = data_act.query('platform == "XOne"').dropna()
# применение функции для Xbox One
plot_corr(XOne)
```

```python
# создание среза для платформы Nintendo 3DS с непустыми значениями оценок
DS3 = data_act.query('platform == "3DS"').dropna()
# применение функции для Nintendo 3DS
plot_corr(DS3)
```

Для всех трех платформ получен примерно одинаковый результат: чем выше оценки критиков, тем выше продажи. Можно заметить, что у Nintendo 3DS в отличие от остальных наблюдается слабая прямая связь и с оценками пользователей. Возможно, это связано с более маленькой выборкой, а возможно, при оценке игр этой платформы критики и игроки более единодушны. 

Посмотрим, действительно ли это так, сравнив корреляции между оценками критиков и пользователей для всех трех платформ.

```python
# вывод коэффициентов корреляции 
print(f"PlayStation 4: {PS4['user_score'].corr(PS4['critic_score']).round(3)}")
print(f"Xbox One: {XOne['user_score'].corr(XOne['critic_score']).round(3)}")
print(f"Nintendo 3DS: {DS3['user_score'].corr(DS3['critic_score']).round(3)}")
```

## Составление портрета пользователя каждого региона

Создадим функции, позволяющие по заданным параметрам быстро делать сводные таблицы и строить круговые диаграммы. 

```python
# функция для создания сводной таблицы по выбранному параметру
def portret_by_region(parameter):
    return data_act.pivot_table(index=parameter, 
                                values=['na_sales', 'eu_sales', 'jp_sales', 'other_sales'], 
                                aggfunc='sum')
```

```python
# функция для построения круговой диаграммы
def top5_pie(parameter, reg, title, ax):
    
    # выбор топ-5 из сводной таблицы
    top5 = (portret
            .reset_index()
            .sort_values(by=reg, ascending=False)
            .reset_index(drop=True)
            .loc[:4,[parameter,reg]])
    
    # выбор категорий, не вошедших в топ-5
    other = (portret
            .reset_index()
            .sort_values(by=reg, ascending=False)
            .reset_index(drop=True)
            .loc[5:,[parameter,reg]])
    
    # добавление в топ-5 суммы категорий, не вошедших в топ-5
    if other[reg].sum() > 0.0:
        top5 = top5.append({parameter:'other', reg:other[reg].sum()}, ignore_index=True)
    
    # построение круговой диаграммы
    ax1 = fig.add_subplot(ax)
    plt.pie(data=top5, 
            x=reg, 
            labels=parameter, 
            autopct='%.1f%%',
            colors=sns.color_palette("Set2"))
    plt.title(title, fontsize=15)
```

### Самые популярные платформы в каждом регионе

Для начала посмотрим, какие платформы популярны в разных регионах.

```python
# вывод сводной таблицы
portret = portret_by_region('platform')
portret.style.background_gradient('Blues')
```

```python
# построение круговых диаграмм
fig = plt.figure(figsize=(12,12))

top5_pie('platform', 'eu_sales', 'продажи в Европе', 221)
top5_pie('platform', 'jp_sales', 'продажи в Японии', 222)
top5_pie('platform', 'na_sales', 'продажи в Северной Америке', 223)
top5_pie('platform', 'other_sales', 'продажи в других странах', 224)

plt.show()
```

Можно заметить, что в каждом регионе свои лидеры среди платформ. Так, в Японии львиную долю рынка занимает Nintendo 3DS - 45,6%, затем идут PlayStation 3 (18,3%) и PlayStation Vita (10,9%). ## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Проверим, одинаково ли в среднем оценивают пользователи игры на платформах Xbox One и PC. Для начала разделим таблицу на 2 датасета.

```python
XOne = data_act.query('platform == "XOne"').dropna(subset=['user_score']) # таблица с данными для XOne
PC = data_act.query('platform == "PC"').dropna(subset=['user_score']) # таблица с данными для PC
```

Рассчитаем для новых выборок среднее и дисперсию. 

```python
# расчет и вывод средних и дисперсий
print(f"Средняя пользовательская оценка для Xbox One: {XOne['user_score'].mean():.2f}")
print(f"Средняя пользовательская оценка для PC: {PC['user_score'].mean():.2f}")
print(f"Дисперсия оценок пользователей для Xbox One: {XOne['user_score'].var():.2f}")
print(f"Дисперсия оценок пользователей для PC: {PC['user_score'].var():.2f}")
```

Для полученных выборок средние очень близки, а дисперсии различаются. Сформулируем нулевую и альтернативную гипотезы.

Гипотеза H₀ звучит так:

*Средние пользовательские рейтинги платформ Xbox One и PC равны.*

Альтернативная гипотеза H₁:

*Средние пользовательские рейтинги платформ Xbox One и PC различаются.*

Воспользуемся методом для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам. Поскольку на предыдущем шаге мы выяснили, что дисперсии в выборках различаются, укажем параметр `equal_var = False`. Примем критический уровень статистической значимости равным 5%.

```python
# критический уровень статистической значимости
alpha = 0.05 
# применение t-теста
results = st.ttest_ind(XOne['user_score'], PC['user_score'], equal_var = False) 
# вывод p-значения
print('p-значение: ', results.pvalue)
# вывод результатов проверки
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Мы не можем откинуть гипотезу о том, что средние пользовательские рейтинги для Xbox One и PC равны. Таким образом, можно сказать, что пользователи оценивают игры на этих двух платформах в среднем одинаково.

### Средние пользовательские рейтинги жанров Action и Sports разные

Теперь проверим, действительно ли различаются оценки пользователей для жанров Action и Sports. Получим выборки для этих жанров.

```python
action = data_act.query('genre == "Action"').dropna(subset=['user_score']) # таблица с данными для жанра Action 
sports = data_act.query('genre == "Sports"').dropna(subset=['user_score']) # таблица с данными для жанра Sports
```

Рассчитаем среднее и дисперсию. 

```python
# расчет и вывод средних и дисперсий
print(f"Средняя пользовательская оценка для жанра Action: {action['user_score'].mean():.2f}")
print(f"Средняя пользовательская оценка для жанра Sports: {sports['user_score'].mean():.2f}")
print(f"Дисперсия оценок пользователей для жанра Action: {action['user_score'].var():.2f}")
print(f"Дисперсия оценок пользователей для жанра Sports: {sports['user_score'].var():.2f}")
```

Для полученных выборок и дисперсии, и средние различаются. Нулевая гипотеза H₀ будет звучать так:

*Средние пользовательские рейтинги жанров Action и Sports равны.*

Альтернативная гипотеза H₁:

*Средние пользовательские рейтинги жанров Action и Sports различаются.*

Воспользуемся тем же методом, который использовался для проверки предыдущей гипотезы. Примем критический уровень статистической значимости равным 5%.

```python
# критический уровень статистической значимости
alpha = 0.05 
# применение t-теста
results = st.ttest_ind(action['user_score'], sports['user_score'], equal_var = False) 
# вывод p-значения
print('p-значение: ', results.pvalue)
# вывод результатов проверки
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Таким образом, при 5%-ном уровне значимости можно сделать вывод, что средние пользовательские рейтинги жанров Action и Sports различаются. Если судить по выборкам, то жанру Action ставят более высокие оценки.
